On va regarder plus en détail la variable `content_rating`.

```python
data.content_rating.value_counts()
```

Une petite recherche et on trouve que historiquement :
- Approved = M = GP = PG
- X = NC-17
- Not Rated = Unrated = NR

On veut donc remplacer Passed, Approved, M et GP par PG, X par NC-17 et Not Rated et Unrated par NR qui sont les notations utilisées aujourd'hui.

```python
def replace_rating(row):
    if row['content_rating'] in ['Passed', 'Approved', 'M', 'GP']:
        return 'PG'
    elif row['content_rating'] in ['Not Rated', 'Unrated']:
        return 'UR'
    elif row['content_rating'] == 'X':
        return 'NC-17'
    else:
        return row['content_rating']
    
data['content_rating'] = data.apply(replace_rating, axis=1)
data.content_rating.value_counts()
```

On a déjà géré presque tout, le dernier point en suspens est la variable `genre` qu'il faut spliter puis créer des OneHotEncoded variables.

```python
test = data['genres'].apply(lambda row : row.replace('-','').split('|'))
genres = pd.Series(test.sum()).unique()
genres
```

```python
for gen in genres:
    data[gen] = data['genres'].apply(lambda row : gen in row)

data
```

```python
# suppression de la variable d'origine genres
data.drop('genres', axis=1, inplace=True)
data.shape
```
