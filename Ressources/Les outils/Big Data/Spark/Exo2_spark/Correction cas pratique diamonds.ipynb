### Quelques manipulations de bases de données  
L'API dataframes de Spark offre un grand nombre de fonctions natives qui permettent de mettre en oeuvre la pluspart des traitements que l'on a l'habitude de faire avec `pandas`. Vous pouvez notamment trouver un inventaire de ces fonctions dans [cet article](https://hackersandslackers.com/transforming-pyspark-dataframes/). En vous aidant de cette syntaxe, faites le traitement suivant :   
- Dans la table `diamonds`, créez une nouvelle variable `categ_price`, égale à :  
  - "low" si le prix du diamant est inférieur à 1000$, 
  - "medium" si le prix du diamant est inférieur à 3000$,  
  - "high" si le prix du diamant est inférieur à 8000$,
  - "very high" si le prix est au delà.  
 _Indications_ : Vous aurez besoin pour cette question des fonctions `withColumn()`, `when()`, `col()`, et `lit()`.

```
# création d'une nouvelle variable : on entre dans une syntaxe assez spécifique avec notamment
# withColumn() et lit()
# un petit exemple : 
from pyspark.sql.functions import lit
test = diamonds.withColumn('test', lit("texte"))
display(test.select("test"))
```

```
# Pour notre cas on fait quelque chose de plus complexe
from pyspark.sql.functions import lit, when, col
diamonds = diamonds.withColumn('categ_price', when(col("price") <= 1000, lit("0. low")).when(col("price") <= 3000, lit("1. medium")).when(col("price") <= 8000, lit("2. high")).otherwise(lit("3. very high")))
display(diamonds.select("price", "categ_price"))
```

```
# On peut vérifier aussi en repassant par pandas : 
import pandas as pd
pandas_diamonds = diamonds.toPandas()
print(pandas_diamonds.loc[pandas_diamonds.price >= 8000, ["price", "categ_price"]])
```

### SQL  
Databricks vous offre aussi la possibilité de coder directement en SQL.

On utilise `%sql` our coder directement en SQL :

```
%sql
/* On affiche les 10 premières lignes après avoir importé la table */
DROP TABLE IF EXISTS diamonds_sql; 

CREATE TABLE diamonds_sql
USING csv
OPTIONS (path "/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv", header "true", inferSchema "true");

SELECT * from diamonds_sql
LIMIT 10
```

```
%sql
/* On crée la variable avec des case when */ 
DROP TABLE IF EXISTS diamonds_sql_categ;
CREATE TABLE diamonds_sql_categ AS
SELECT *, 
CASE    
        WHEN price <= 1000 THEN 'low'
        WHEN price <= 3000 THEN 'medium'
        WHEN price <= 8000 THEN 'high'
        ELSE 'very high'
    END AS categ_price
FROM diamonds_sql;

SELECT price, categ_price FROM diamonds_sql_categ
```

```
# On crée notre dataframe avec les features et le y 
# on regroupe toutes les variables numériques avec VectorAssembler
from pyspark.ml.feature import VectorAssembler
name_num = diamonds_num.drop("_c0", "color", "clarity", "cut", "price", "categ_price").columns
vectorAssembler = VectorAssembler(inputCols = name_num, outputCol = 'features')

diamonds_reg = vectorAssembler.transform(diamonds_num)
diamonds_reg.select(['features', 'price']).show(20)
```

```
# On fait tourner le modèle de régression linéaire
from pyspark.ml.regression import LinearRegression
lr = LinearRegression(featuresCol = 'features', labelCol='price', maxIter=100)
lr_model = lr.fit(diamonds_reg)
```

```
# on sort le R2
lr_summary = lr_model.summary
print(lr_summary.r2)
```

```
# on affiche les coefficients
print(lr_model.coefficients)
```

Il manque la constante dans le vecteur `lr_model.coefficients`, on la récupère via `lr_model.intercept`. On va récupérer aussi les p-values et afficher tout ça joliment.

```
#on récupère les coefficients
liste_coefs = list(lr_model.coefficients)
liste_coefs.append(lr_model.intercept)
var = name_num.copy()
var.append('cste')
coefs = pd.DataFrame(data=liste_coefs, index=var, columns=["coefficients"])

# puis les p-values
coefs["pvalues"] = lr_summary.pValues
coefs
```

On déduit des p-values qu'on peut sortir y et z

```
name_num = diamonds_num.drop("_c0", "color", "clarity", "cut", "price", "categ_price", "y", "z").columns
vectorAssembler = VectorAssembler(inputCols = name_num, outputCol = 'features')
diamonds_reg = vectorAssembler.transform(diamonds_num)
```

```
# On fait maintenant un modèle de prédiction
splits = diamonds_reg.randomSplit([0.7, 0.3])
train_diamonds = splits[0]
test_diamonds = splits[1]

lr_train_test = lr.fit(train_diamonds)
print(lr_train_test.summary.r2)
```

```
# on applique la prédiction
lr_predictions = lr_train_test.transform(test_diamonds)
display(lr_predictions)
```

```
# on sort MAE et rmse pour juger de la qualité de la prédiction
test_result = lr_train_test.evaluate(test_diamonds)
# RMSE
print(test_result.rootMeanSquaredError)
# MAE
print(test_result.meanAbsoluteError)
```

### Classification  
- Entraînez un modèle randomforest qui permette de classer le diamant en `categ_price` en fonction de ses caractéristiques. Proposez des indicateurs pour évaluer la qualité prédictive de votre modèle.   

__Indications__  
Pour gérer les variables catégorielles pour la classification, deux approches sont possibles et évidemment, il va falloir faire les deux :
>- La première approche consiste à considérer les variables catégoriques comme étant **non-ordonnées** :
>>1. transformez-les en index avec `StringIndexer` (vous devrez aussi utiliser cette fonction pour votre variable expliquée!)
>>2. créez votre colonne "features" mais au sein de celle-ci vous devrez ensuite appliquer un  `VectorIndexer` pour indiquer quelles sont les variables catégorielles (fixez bien le paramètre `maxCategories`)
>- La seconde approche consiste à considérer les variables catégoriques comme étant **ordonnées** :
>>1. transformez-les en index avec `StringIndexer` si vous pouvez, seulement ce sera pas toujours possible...
>>2. créez votre colonne "features" sans avoir besoin d'appliquer un `VectorIndexer` puisque, une fois transformées, les variables catégorielles sont considérées comme des variables numériques dans cette approche

```
display(diamonds)
```

```
# On peut dans ce cas garder les variables catégoriques en les passant par StringIndexer : 
from pyspark.ml.feature import StringIndexer, VectorAssembler, VectorIndexer
diamonds_rf = diamonds.drop("_c0", "price")

indexer = StringIndexer(inputCol="color", outputCol="color_index").fit(diamonds_rf)
diamonds_rf = indexer.transform(diamonds_rf)
indexer = StringIndexer(inputCol="clarity", outputCol="clarity_index").fit(diamonds_rf)
diamonds_rf = indexer.transform(diamonds_rf)
indexer = StringIndexer(inputCol="cut", outputCol="cut_index").fit(diamonds_rf)
diamonds_rf = indexer.transform(diamonds_rf)

# on le fait aussi pour categ_price
indexer_pred = StringIndexer(inputCol="categ_price", outputCol="categ_price_index").fit(diamonds_rf)
diamonds_rf = indexer_pred.transform(diamonds_rf)

diamonds_rf = diamonds_rf.drop("color", "clarity", "cut")

# On assemble dans features
vectorAssembler = VectorAssembler(inputCols = diamonds_rf.drop("categ_price_index", "categ_price").columns, outputCol = 'features')
diamonds_rf = vectorAssembler.transform(diamonds_rf).select("features", "categ_price_index", "categ_price")

# Attention : on doit spécifier quelles sont les variables catégoriques dans features avec VectorIndexer!
featureIndexer = VectorIndexer(inputCol="features", outputCol="index_features", maxCategories=9).fit(diamonds_rf)
diamonds_rf = featureIndexer.transform(diamonds_rf)

#display(diamonds_rf.limit(20))

# On sépare en traînement et test
(train_rf, test_rf) = diamonds_rf.randomSplit([0.7, 0.3])
```

```
# Après tout ça, on peut entraîner notre modèle sur les mêmes bases partitionnées en entraînement et test : 
from pyspark.ml.classification import RandomForestClassifier
rf = RandomForestClassifier(labelCol="categ_price_index", featuresCol="index_features", numTrees=50)

modele_rf = rf.fit(train_rf)
predict_diamonds = modele_rf.transform(test_rf)
predict_diamonds.show(5)
```

```
# on peut repasser les prédictions avec leurs labels 
from pyspark.ml.feature import IndexToString

label_converter = IndexToString(inputCol="prediction", outputCol="predictedLabel",
                               labels=indexer_pred.labels)
predict_diamonds_lab = label_converter.transform(predict_diamonds)
predict_diamonds_lab.show()
```

```
# on évalue la qualité du modèle
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
evaluator = MulticlassClassificationEvaluator(
  labelCol="categ_price_index", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predict_diamonds_lab)
print(accuracy)
```

```
# Matrice de confusion à la main
display(predict_diamonds_lab.stat.crosstab("categ_price", "predictedLabel").orderBy("categ_price_predictedLabel"))
```

On va maintenant regarder la 2ème approche qui consiste à considérer que les variables catégoriques sont ordonnées et donc on peut se contenter de les transformer en valeurs numériques sans ensuite utiliser le VectorIndexer. La difficulté réside dans le passage en valeurs numériques car on ne peut plus utiliser directement StringIndexer. Du moins pas tel quel...

```
# Nouvelle approche pour les variables catégoriques :
from pyspark.ml.feature import StringIndexer, VectorIndexer
diamonds_rf = diamonds.drop("_c0", "price")

# Pour color, on peut simplement préciser dans StringIndexer que les labels sont rangés par ordre alphabétique décroissant
# puisque les modalités sont ['J','I','H','G','F','E','D'] du pire au meilleur
indexer = StringIndexer(inputCol="color", outputCol="color_index", stringOrderType="alphabetAsc").fit(diamonds_rf) #
diamonds_rf = indexer.transform(diamonds_rf)

# Pour les autres variables, l'ordre des labels n'est pas alphabétique, on peut donc pas utiliser simplement le StringIndexer
# Plusieurs méthodes possibles, là j'utilise les user-defined fonctions pour créer les indexes
from pyspark.sql.functions import udf
@udf('double')
def chg_cut(s):
  labels = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
  if s in labels:
    return float(labels.index(s))
  
@udf('double')
def chg_clarity(s):
  labels = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
  if s in labels:
    return float(labels.index(s))
    
diamonds_rf = diamonds_rf.withColumn("cut_index", chg_cut("cut")).withColumn("clarity_index", chg_clarity("clarity"))


# on passe aussi en index pour categ_price et pour changer on utiliser l'ordre alphabétique croissant
indexer_pred = StringIndexer(inputCol="categ_price", outputCol="categ_price_index", stringOrderType="alphabetAsc").fit(diamonds_rf)
diamonds_rf = indexer_pred.transform(diamonds_rf)

diamonds_rf.show()
```

```
# Pour la suite on reprend exactement comme précédemment
# On supprime les colonnes originales
diamonds_rf = diamonds_rf.drop("color", "clarity", "cut", "y", "z")

# On assemble dans features (et on a plus besoin de spécifier quelles sont les variables catégoriques)
from pyspark.ml.feature import VectorAssembler
vectorAssembler = VectorAssembler(inputCols = diamonds_rf.drop("categ_price_index", "categ_price").columns, outputCol = 'features')
diamonds_rf = vectorAssembler.transform(diamonds_rf).select("features", "categ_price_index", "categ_price")

# On sépare en traînement et test
(train_rf, test_rf) = diamonds_rf.randomSplit([0.7, 0.3])

# On entraîne notre modèle
rf = RandomForestClassifier(labelCol="categ_price_index", featuresCol="features", numTrees=50)
modele_rf = rf.fit(train_rf)
predict_diamonds = modele_rf.transform(test_rf)

# on évalue la qualité du modèle (accuracy et matrice de confusion)
evaluator = MulticlassClassificationEvaluator(
  labelCol="categ_price_index", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predict_diamonds)

print(accuracy)

predict_diamonds.stat.crosstab("categ_price_index", "prediction").orderBy("categ_price_index_prediction").show()
```
