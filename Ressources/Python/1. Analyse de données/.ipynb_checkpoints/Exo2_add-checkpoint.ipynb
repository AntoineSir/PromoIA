## Convertir l'âge en une valeur numérique

## Convertir les revenus en valeurs numériques

Appliquer exactement la même méthodologie pour convertir le revenu (colonne 'How much total combined money did all members of your HOUSEHOLD earn last year?') en valeur numérique.  
Nommer la nouvelle variable 'income_num'.

```python
def rev(ch):
    if pd.isnull(ch) or ch[0]=='P':
        return None
    else:
        ch = ch.replace('$','')
        ch = ch.replace(',','')
        ch = ch.replace(' and up','')
        liste = ch.split(' ')
        inf = int(liste[0])
        sup = int(liste[-1])
        return (inf+sup)/2
```

```python
data['income_num']=data['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(rev)
data['income_num'].describe()
```
