import time

import urllib.request

def download_progress(count, size, total):
  global start_time, previous_state 
  
  if count == 0:
    start_time = time.time()
    previous_state = 0
  duration = time.time() - start_time  
  
  progress = count*size/(1024*1024)
  
  #clear_output(wait=True)
  if progress - previous_state > 0.5:
    previous_state = progress
    print("{:.1f} MB téléchargés en {:.1f} secondes".format(progress, duration))

download_url = "https://public.opendatasoft.com/explore/dataset/airbnb-listings/download/?format=csv&timezone=Europe/Berlin&lang=fr&use_labels_for_header=true&csv_separator=%3B"

On peut utiliser une fonction udf
``` Python
# on crée fonction udf pour faire l'opération sur la colonne
@udf()
def del_r_eol(s): #le nom chelou c'est pour delete \r at end-of-line
  return s.replace('\r','')

# puis on crée un nouvelle colonne à partir de cette fonction et la renommant correctement au passage et on supprime l'ancienne
df = df.withColumn("Features", del_r_eol("Features\r")).drop("Features\r")
```

````
from pyspark.sql.functions import *

# autre méthode sans udf
df = df.withColumn('Features', regexp_replace('Features\r', '\r', '')).drop('Features\r')

# on vérifie
print(df.columns, df.limit(20).select(df.columns[-1]).collect(), sep='\n\n')
````

````
a = "I am a big out doors fan."
b = "810552"

#on crée un nouveau df Spark sans les 2 lignes problématiques 
df_wo_pbm_lines = df.filter(~col("ID").isin([a, b]))

# on pourrait en rester là, mais pour faire ça bien on peut regarder de plus près et voir s'il y a pas moyen de corriger ces lignes
# pour ça, on crée un dataframe Pandas afin de manipuler ces lignes plus facilement
pbm_line = df.filter(col("ID").isin([a, b])).toPandas()
````

````
# pour faire nos modifs et nos tests, on copie le df pandas histoire de pas avoir à le recréer à chaque fois qu'on ferait une mauvaise manip
pbm_line_mod = pbm_line.copy()

# on récupère la variable "Features" qui est à la fin
pbm_line_mod.loc[0,"Features"] = pbm_line_mod.loc[1,"Features"]

# on modifie la variable "host about" en ajoutant ce qui a été coupé et qui se situe dans "ID" de la 2ème ligne
pbm_line_mod.loc[0,"Host About"] += " "+pbm_line_mod.loc[1,"ID"]

##### "plus qu'à" récupérer les informations de la 2ème ligne pour les coller dans la 1ère
liste = list(pbm_line_mod.loc[1].values)

# on supprime les 't' et 'f' qui correspondent à des erreurs dans le scraping initial
while 't' in liste:
  liste.remove('t')
  
while 'f' in liste:
  liste.remove('f')

# si on remplace telles quelles les données à partir de "Host About":
# on s'aperçoit qu'il y a un décalage de 2 colonnes à partir de "Weekly Price"
# puis à nouveau un décalage de 1 colonne supplémentaire à partir de "Availability 90"

idx_host_about = list(pbm_line_mod.columns).index("Host About")
idx_amenities = list(pbm_line_mod.columns).index("Amenities")
idx_wkl_price = list(pbm_line_mod.columns).index("Weekly Price")
idx_avail90 = list(pbm_line_mod.columns).index("Availability 90")
idx_rvw_mth = list(pbm_line_mod.columns).index("Reviews per Month")

for i in range(idx_host_about+1, idx_amenities+1):
  pbm_line_mod.iloc[0,i] = liste[i-idx_host_about]

for i in range(idx_wkl_price, idx_avail90):
  pbm_line_mod.iloc[0,i] = liste[i-idx_host_about-2] #-2 il y a 2 colonnes de décalage
  
for i in range(idx_avail90+1, idx_rvw_mth+1):
  pbm_line_mod.iloc[0,i] = liste[i-idx_host_about-3]  #-3 car il y a une nouvelle colonne de décalage

# on peut fixer Availability 90 à 0 car dernière màj du calendrier à plus de 4 mois
pbm_line_mod.loc[0,"Availability 90"] = 0

# on recréé Geolocation avec les bonnes latitudes/longitudes
pbm_line_mod.loc[0,"Geolocation"] = "{},{}".format(pbm_line_mod.loc[0,"Latitude"], pbm_line_mod.loc[0,"Longitude"])
````

````
# pour pouvoir afficher toutes les colonnes on repasse par un dataframe spark
# avec la petite subtilité qu'on convertit en str pour pas avoir de problème de types car spark et numpy ça fait pas bon ménage...
display(spark.createDataFrame(pbm_line_mod.astype(str)))
````

````
# reste à ajouter cette nouvelle ligne dans le df spark
# l'idée c'est de créer un dataframe spark à partir de la (ou les lignes) du df pandas puis on les colle avec union

# le problème c'est les types des variables car le passage à pandas puis le retour à un df spark met le bazar
# c'est du bidouillage mais pour contourner ça, vu qu'il y a qu'une seule ligne, on fait un intermédiaire avec un csv pour récupérer le schéma

# d'abord il faut modifier un peu le dataframe pandas en mettant 2 fois la même ligne pour avoir bien le même type de variables dans chaque colonne
pbm_line_mod.loc[1] = pbm_line_mod.loc[0]
# rmq : on met 2 lignes car si on en prend une seule, on passe à un objet Series plutôt que dataframe, ce qui met aussi le bordel

#on fait un csv avec ce dataframe qui a 2 fois la même ligne
pbm_line_mod.to_csv('/tmp/temp.csv', header=True, sep=";")

#on déplace le csv dans dbfs
dbutils.fs.mv("file:/tmp/temp.csv","dbfs:/data/temp.csv")

#on importe ce csv
new_line = spark.read.csv("dbfs:/data/temp.csv", header=True, inferSchema=True, sep=";", escape="\"", multiLine=True)

#on modifie le df spark pour ne garder qu'une seule ligne et supprimer la colonne _c0 qui correspond à l'index du dataframe pandas
new_line = new_line.filter(col("_c0")==0).drop("_c0")
````

````
#puis il reste plus qu'à "coller" les dataframes spark
df_clean = df_wo_pbm_lines.union(new_line)
display(df_clean)
````
