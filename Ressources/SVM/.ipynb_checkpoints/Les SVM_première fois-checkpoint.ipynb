# Les SVM : première rencontre

## 1 - Cas linéaire

```python
import numpy as np
import matplotlib.pyplot as plt
```

- Simuler un ensemble d’entraînement de dimension 2 linéairement séparable, c'est-à-dire que l'on peut séparer par une droite. Il devra être composé de 100 observations avec 2 caractéristiques chacune (*indication : on peut prendre par exemple y=1 pour x1 > 0.5*)
- Afficher cet ensemble en colorant par label.

```python
np.random.seed(0)
X = np.random.rand(100, 2)
x0 = X[:,0]
x1 = X[:,1]
#y = x1 > 0.5
y = x1 > 0.5*x0+0.2
plt.scatter(x0,x1,c=y);
```

- Entrainer un classifieur SVM linéaire (LinearSVC) et calculer l'accuracy

```python
from sklearn.svm import LinearSVC
#from slkearn.metrics import accuracy_score
svc = LinearSVC()
svc.fit(X,y)
svc.score(X,y) #accuracy_score(svc.predict(X),y)
```

- Faire varier le paramètre C pour voir son effet sur le modèle 

- Faire un graphique avec l'ensemble de décision en utilisant une fonction `frontiere` à définir (le graphique doit donc représenter les points et les 2 espaces séparés par la frontière de décision)

```python
def frontiere(model,ax=None):
    if ax == None:
        ax = plt.gca()

    # délimitation des axes
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # création d'une grille pour évaluer le modèle
    x = np.linspace(xlim[0], xlim[1], 1000) #(30,1)
    y = np.linspace(ylim[0], ylim[1], 1000) #(30,1)

    xx, yy = np.meshgrid(y, x) #2 fois (30,30)

    xy = np.vstack([xx.ravel(), yy.ravel()]).T #(900,2)

    z = model.predict(xy) #(900,1)
    z = z.reshape(xx.shape) #(30,30)

    d = model.decision_function(xy).reshape(xx.shape) #(900,1) puis (30,30)


    # afficher la frontière et les marges
    ax.contour(xx, yy, d, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])

    ax.contourf(xx, yy, z, cmap = plt.cm.Paired, alpha=0.5)

    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
fig = plt.figure(figsize=(20,20))
idx_plt=1
for C in [10**k for k in range(-6,6)]:
    axi = fig.add_subplot(4,3,idx_plt)
    model = LinearSVC(C=C)
    model.fit(X,y)
    frontiere(model, ax=axi)
    plt.scatter(x0,x1,c=y)
    axi.set_title(f"pour C = {C} l'accuracy est de {model.score(X,y)}", fontsize=14)
    idx_plt += 1
```

- Sur cet ensemble, ajouter du bruit aux y (en faisant y = 1 - y pour 10 lignes choisies au hasard).
- Afficher l'ensemble, entrainer un SVM linéaire (LinearSVC) et faire varier le paramètre C.

```python
np.random.seed(0)
X = np.random.rand(100, 2)
x0 = X[:,0]
x1 = X[:,1]
#y = x1 > 0.5
y = x0 > 0.5
#lignes_bruitees = list(range(10))
lignes_bruitees = np.random.choice(len(y),10)
y[lignes_bruitees]=1-y[lignes_bruitees]
plt.scatter(x0,x1,c=y);
```

```python
fig = plt.figure(figsize=(20,20))
idx_plt=1
for C in [10**k for k in range(-6,6)]:
    axi = fig.add_subplot(4,3,idx_plt)
    model = LinearSVC(C=C)
    model.fit(X,y)
    frontiere(model, ax=axi)
    plt.scatter(x0,x1,c=y)
    axi.set_title(f"pour C = {C} l'accuracy est de {model.score(X,y)}", fontsize=14)
    idx_plt += 1
```
