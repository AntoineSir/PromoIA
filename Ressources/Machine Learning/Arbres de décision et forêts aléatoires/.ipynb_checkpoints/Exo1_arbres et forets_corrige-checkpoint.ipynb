On va utiliser le jeu de données Pima Indian Diabetes disponible [là par exemple](https://www.kaggle.com/uciml/pima-indians-diabetes-database). Le dataset est déjà chargé dans l'environnement.

Problème des variables ayant des valeurs à 0. On pouvait remplacer par la moyenne/médiane mais comme plusieurs d'entre vous l'ont fait remarqué, c'est pas idéal...

Pour aller un peu plus loin, on peut déjà remarquer que pour Glucose, BloodPressure et BMI, cela concerne peu d'individus donc on peut choisir de s'en débarrasser.

Dans un second temps, on peut tester avec SkinThickness mais celle-ci n'est quasiment pas du tout corrélée avec l'Outcome. Et on se rend compte lorsqu'on la conserve que l'importance de cette variable est très faible, on peut donc choisir de s'en débarrasser.

Reste,la question de l'insuline qui est en vérité assez problématique car il y a vraiment beaucoup de valeurs manquantes (plus de 50%...). En réalité, dans un cas réel, il faudrait vraiment aborder cette question avec le personnel médical derrière cette étude. Ici, aucun solution n'étant vraiment optimale, on peut choisir de s'en débarrasser ou bien d'imputer la moyenne/médiane.

```python
data = data[(data.BloodPressure != 0) & (data.BMI != 0) & (data.Glucose != 0)] #suppression des individus avec BloodPressure, BMI et Glucose à 0
data.drop(["SkinThickness","Insulin"], axis=1, inplace=True) #suppresion de la colonne SkinThickness
data.shape
```

## Préparation des différents datasets pour la modélisation

```python
X = data.drop("Outcome", axis=1)
y = data["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## Un premier arbre de décision

Créer et entrainer un arbre de décision puis calculer l'accuracy et afficher l'arbre obtenu.

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)

# mesure de l'accuracy
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# affichage de l'arbre
from sklearn.tree import export_graphviz
from IPython.display import Image  
import pydotplus

dot_data = export_graphviz(clf, out_file=None,  
                filled=True, rounded=True,
                special_characters=True,
                feature_names = X.columns,
                class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data)  
graph.write_png('diabetes.png')
Image(graph.create_png())
```

```python
# affichage de l'importance des features
imp = pd.DataFrame(clf.feature_importances_, columns=['Feature importance'], index=X.columns)
imp.sort_values('Feature importance', ascending=False)
```

## Un second arbre de décision

Refaire la même chose en modifiant les paramètres de la classe `DecisionTreeClassifier`, notamment `max_depth` et `criterion`.

```python
# Modèle
clf = DecisionTreeClassifier(criterion = 'entropy', max_depth=3)
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)

# Accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# Affichage
dot_data = export_graphviz(clf, out_file=None,  
                filled=True, rounded=True,
                special_characters=True,
                feature_names = X.columns,
                class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data)  
graph.write_png('diabetes.png')
Image(graph.create_png())
```

## Un peu de validation croisée pour affiner les hyperparamètres

Utiliser la fonction adéquate pour tester différentes valeurs des paramètres(`max_depth` et `criterion`).

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'max_depth': list(range(1,11)), 'criterion':["gini", "entropy"]}

grid_cv = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=50, scoring='accuracy')
grid_cv.fit(X_train, y_train)
```

```python
print("Meilleurs paramètres sur le jeu d'entraînement:")
print(grid_cv.best_params_)

# Afficher les performances correspondantes
print("Résultats de la validation croisée :")
for mean, std, params in zip(
        grid_cv.cv_results_['mean_test_score'],
        grid_cv.cv_results_['std_test_score'],
        grid_cv.cv_results_['params']
    ):

    print("'accuracy' = {:.3f} (+/-{:.03f}) for {}".format(mean,std*2,params))
```

```python
# autre affichage
res = pd.DataFrame(zip(grid_cv.cv_results_['mean_test_score'],
                       grid_cv.cv_results_['std_test_score'],
                       grid_cv.cv_results_['params']),
                  columns = ['mean', 'std', 'params'])
res.head()
```

```python
# avec les courbes d'évolution de l'accuracy
plt.plot(np.arange(10), res.loc[:9,'mean'], label='Gini')
plt.plot(np.arange(10), res.loc[10:,'mean'], label='Entropy')
plt.legend();
```

```python
best_tree = grid_cv.best_estimator_
```

```python
y_pred = best_tree.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

## Et une petite forêt aléatoire pour finir...

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=250)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

## ...avec validation croisée pour le nombre d'arbres de la forêt

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'n_estimators':[10, 100, 250, 500, 1000]}

grid_cv = GridSearchCV(RandomForestClassifier(), param_grid, cv=40, scoring='accuracy')
grid_cv.fit(X_train, y_train)
```

```python
print("Meilleurs paramètres sur le jeu d'entraînement:")
print(grid_cv.best_params_)

# Afficher les performances correspondantes
print("Résultats de la validation croisée :")
for mean, std, params in zip(
        grid_cv.cv_results_['mean_test_score'],
        grid_cv.cv_results_['std_test_score'],
        grid_cv.cv_results_['params']
    ):

    print("'accuracy' = {:.3f} (+/-{:.03f}) for {}".format(mean,std*2,params))
```

```python
best_rf = grid_cv.best_estimator_
y_pred = best_rf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# affichage de l'importance des features
imp = pd.DataFrame(best_rf.feature_importances_, columns=['Feature importance'], index=X.columns)
imp.sort_values('Feature importance', ascending=False)
```
