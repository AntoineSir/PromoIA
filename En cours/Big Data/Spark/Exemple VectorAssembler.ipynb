Exemple qui vient de la [documentation](https://spark.apache.org/docs/latest/ml-features#vectorassembler)

```
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler

dataset = spark.createDataFrame(
    [(0, 18, 1.0, 2.5, 1.0, Vectors.dense([0.0, 10.0, 0.5])),
     (1, 12, 3.0, 2.2, 5.0, Vectors.dense([1.5, 8.0, 0.7])),
     (2, 11, 4.0, 2.3, 2.0, Vectors.dense([0.0, 2.5, 0.5])),
     (3, 13, 1.5, 2.4, 6.0, Vectors.dense([0.4, 10.0, 1.5]))],
    ["id", "var1", "var2", "var3", "var4", "var5"])
dataset.show()
```

```
assembler = VectorAssembler(
    inputCols=["var1", "var2", "var3", "var4", "var5"],
    outputCol="features")

output = assembler.transform(dataset)
output.show()
```
